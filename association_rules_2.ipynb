```python
subset = subset.fillna(5)
```

```python
te = TransactionEncoder()
item_set = te.fit_transform(subset.values)
```

```python
item_set_df = pd.DataFrame(data=item_set, columns=te.columns_)
item_set_df
```

```python
sets = fpgrowth(
    item_set_df, 
    min_support=0.001, 
    use_colnames=True, 
)
```

```python
sets
```

```python
rules = association_rules(sets, metric='confidence', min_threshold=0.5)
```

```python
rules
```

```python
rules.to_csv('pravila_asocijacije.csv', index=False)
```

```python
sorted_rules = rules.sort_values(by='lift', ascending=False)
```

```python
for r in sorted_rules.values:
    print(r[:2], f'L={r[6]}')
    print(' ')
```
