```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
df = pd.read_csv('podaciZaPajtonPrvaAnaliza.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df['INDEKS'] = df['INDEKS'].apply(lambda x: f'INDKES={x}')
df['IDPREDMETA'] = df['IDPREDMETA'].apply(lambda x: f'IDPREDMETA={x}')
df[' DOBIJENAOCENA'] = df[' DOBIJENAOCENA'].apply(lambda x: f'DOBIJENAOCENA={x}')
df['OZNAKAROKA'] = df['OZNAKAROKA'].apply(lambda x: f'OZNAKAROKA={x}')
df['KOLIKOPUTAJEUPISANPREDMET'] = df['KOLIKOPUTAJEUPISANPREDMET'].apply(lambda x: f'KOLIKOPUTAJEUPISANPREDMET={x}')
df
```

```python
subset = df.iloc[:,[1,2,4,5]]
```

```python
subset
```

```python
subset = subset.astype(str)
```

```python
subset.values
```

```python
te = TransactionEncoder()
item_set = te.fit_transform(subset.values)
```

```python
item_set_df = pd.DataFrame(data=item_set, columns=te.columns_)
item_set_df
```

```python
sets = fpgrowth(
    item_set_df, 
    min_support=0.001, 
    use_colnames=True, 
)
```

```python
sets
```

```python
rules = association_rules(sets, metric='lift', min_threshold=0.5)
rules
```

```python
sorted_rules = rules.sort_values(by='lift', ascending=False)
for r in sorted_rules.values:
    print(r[:2], f'L={r[6]}')
    print(' ')
```
